```python
## Pandas is a fast & flexible library for data analysis and data manipulation built on top of Python 

# Import dependencies 

import pandas as pd
```

```python
## Pandas can read data from several file types, including CSV, JSON & SQL


# Read data from csv file by using read_csv function 

pd.read_csv('Desktop/stocks/MSFT.csv')
```

```python
# Dataframes are a datatype used in Pandas, they are a two-dimensional labelled data structure
# Columns within a Dataframe can contain different data types, similar to an Excel file or SQL table

# Assign the data within the CSV file to a variable 

DataFrame = pd.read_csv('Desktop/stocks/MSFT.csv')
```

```python
# Return top 5 rows by using the .head() function | .tail() will return the bottom 5 rows

DataFrame.head()
```

```python
# Add new column into DataFrame for the stock ticker symbol as this was not included in original data set 

DataFrame['Ticker'] = 'MSFT'
```

```python
# Calculate relative change between Open and Close price 

# Calculate delta
relativeChange = DataFrame['Close'] - DataFrame['Open']

# Delta / Original Value (Open price) * 100 
DataFrame['Relative Price Change'] = relativeChange / DataFrame['Open'] * 1

# Format as percentage 
DataFrame['Relative Price Change'] = DataFrame['Relative Price Change'].map('{:.2%}'.format)

# Print result 
DataFrame
```
