```python
## Pandas is a fast & flexible library for data analysis and data manipulation built on top of Python 

# Import dependencies 

import pandas as pd
```

```python
## Pandas can read data from several file types, including CSV, JSON & SQL


# Read data from csv file by using read_csv function 

pd.read_csv('Desktop/data.csv')
```

```python
# Dataframes are a datatype used in Pandas, they are a two-dimensional labelled data structure
# Columns within a Dataframe can contain different data types, similar to an Excel file or SQL table

# Assign the data within the CSV file to a variable 

DataFrame = pd.read_csv('Desktop/data.csv')
```

```python
# Return top 5 rows by using the .head() function | .tail() will return the bottom 5 rows

DataFrame.head()
```

```python
# Sort DataFrame column in descending order based on spend amount:

sortedDataFrame = DataFrame.sort_values('SPEND', ascending=False)

sortedDataFrame
```

```python
# Adding a new column into the existing DataFrame which will store Spend + GST data 

sortedDataFrame['SPEND + GST'] = ''

sortedDataFrame
```

```python
## Looping through DataFrame, multiplying each value in Spend column by 1.1 to calculate Spend + GST

# Create empty list to store the calculated values 

spendPlusGST = []

# Iterate through DataFrame SPEND column and multiply each value by 1.1, append calculated values to empty list 

for x in sortedDataFrame['SPEND']:
        spendPlusGST.append(x*1.1)

# Add spendPlusGST list to existing DataFrame         

sortedDataFrame['SPEND + GST'] = spendPlusGST

# Print results 

sortedDataFrame
```

```python
# Access specific data within the DataFrame by using the .loc function and setting the lookup criteria 

specificLookup = sortedDataFrame.loc[sortedDataFrame['ID']==273]

specificLookup
```

```python
# .loc can also be used to access data with conditional lookups 

conditionalLookup = sortedDataFrame.loc[sortedDataFrame['SPEND'] > 352]

conditionalLookup
```
